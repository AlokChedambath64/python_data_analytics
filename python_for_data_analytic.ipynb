```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=value_counts)
plt.title('Box Plot of Label Occurrences')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()

# 2. Line Plot
plt.figure(figsize=(8, 6))
value_counts.plot(kind='line', marker='o')
plt.title('Line Plot of Label Occurrences')
plt.ylabel('Counts')
plt.xlabel('Labels')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 3. Bar Plot
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar')
plt.title('Bar Plot of Label Occurrences')
plt.ylabel('Counts')
plt.xlabel('Labels')
plt.xticks(rotation=45)
plt.bar_label(value_counts, label_type='center')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

value_counts = value_counts.T

# 1. Scatterplot
plt.figure(figsize=(8, 6))
for col in value_counts.columns:
    plt.scatter(value_counts.index, value_counts[col], label=f'Value {col}')
plt.title('Scatterplot of Label Occurrences')
plt.xlabel('Labels')
plt.ylabel('Counts')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 2. Horizontal Bar Graph
plt.figure(figsize=(8, 6))
value_counts.plot(kind='barh', stacked=False)
plt.title('Horizontal Bar Graph of Label Occurrences')
plt.xlabel('Counts')
plt.ylabel('Labels')
plt.show()

# 3. Stacked Graph
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Graph of Label Occurrences')
plt.ylabel('Counts')
plt.xlabel('Labels')
plt.xticks(rotation=45)
plt.show()

# 4. Histogram with KDE (Kernel Density Estimate)
plt.figure(figsize=(8, 6))
for col in value_counts.columns:
    sns.histplot(value_counts[col], kde=True, label=f'Value {col}', bins=5)
plt.title('Histogram with KDE for Label Occurrences')
plt.xlabel('Counts')
plt.legend()
plt.show()

# 5. Box Plot (repeated for completeness)
plt.figure(figsize=(8, 6))
sns.boxplot(data=value_counts)
plt.title('Box Plot of Label Occurrences')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()

# 6. Area Chart with Labels
plt.figure(figsize=(8, 6))
value_counts.plot(kind='area', stacked=False)
plt.title('Area Chart of Label Occurrences')
plt.ylabel('Counts')
plt.xlabel('Labels')
plt.xticks(rotation=45)
plt.legend(title="Values")
plt.show()
```

# Regex

```python
import re

def find_all_occurrences(pattern, text):
    return re.findall(pattern, text)

text = "The rain in Spain stays mainly in the plain."
pattern = r"in"
print(find_all_occurrences(pattern, text))  
```

```python
import re

def replace_pattern(pattern, replacement, text):
    return re.sub(pattern, replacement, text)

text = "The rain in Spain stays mainly in the plain."
pattern = r"in"
replacement = "XX"
print(replace_pattern(pattern, replacement, text)) 
```

```python
import re

def check_match(pattern, text):
    return bool(re.match(pattern, text))


text = "Hello, world!"
pattern = r"Hello"
print(check_match(pattern, text))  
```

```python
import re

def extract_first_match(pattern, text):
    match = re.search(pattern, text)
    return match.group() if match else None


text = "The rain in Spain."
pattern = r"rain"
print(extract_first_match(pattern, text))  
```

```python
import re

def split_by_pattern(pattern, text):
    return re.split(pattern, text)

text = "apple,banana;grape|orange"
pattern = r"[,;|]"
print(split_by_pattern(pattern, text))  
```
